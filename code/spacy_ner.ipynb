```python
# spaCy NER Example

import spacy

# Load pre-trained spaCy model
nlp = spacy.load("en_core_web_sm")

# Sample text
text = "Barack Obama was born in Hawaii and was the 44th president of the United States."

# Process text
doc = nlp(text)

# Print detected entities
for ent in doc.ents:
    print(ent.text, "->", ent.label_)
```

```python
# Simple rule-based sentiment (Cell 3)
positive_words = ["good", "great", "amazing", "love", "perfect"]
negative_words = ["bad", "terrible", "poor", "hate", "worst"]

def simple_sentiment(text):
    text_lower = text.lower()
    if any(word in text_lower for word in positive_words):
        return "Positive"
    elif any(word in text_lower for word in negative_words):
        return "Negative"
    else:
        return "Neutral"

# Test it
sample_review = "The product quality was great and I love it!"
print("Sentiment:", simple_sentiment(sample_review))
```
